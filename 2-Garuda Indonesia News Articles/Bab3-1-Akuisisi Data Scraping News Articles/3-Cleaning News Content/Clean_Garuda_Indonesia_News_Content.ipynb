# Load Data

```python
# Replace 'your_file_path.csv' with the actual path to your file in Colab Research
# You can upload your file directly to the Colab environment
file_path = 'data_link_berita_with_content.csv' # Example: If you upload a file named 'my_data.csv'

# You can use pandas to read the data, assuming it's a CSV file
import pandas as pd

try:
  df = pd.read_csv(file_path)
  print(f"Successfully loaded data from {file_path}")
  display(df.head())
except FileNotFoundError:
  print(f"Error: File not found at {file_path}. Please make sure you have uploaded the file.")
except Exception as e:
  print(f"An error occurred: {e}")
```

# Remove empty rows

```python
# Remove rows where the 'konten' column is NaN
df_cleaned = df.dropna(subset=['konten'])

print("Original DataFrame shape:", df.shape)
print("Cleaned DataFrame shape:", df_cleaned.shape)

display(df_cleaned.head())
```

# Remove English titles

```python
!pip install langdetect
from langdetect import detect
import numpy as np

# Function to detect language, handles potential errors
def detect_language(text):
    try:
        # Ensure the input is a string
        if isinstance(text, str):
            return detect(text)
        else:
            return 'unknown' # Or handle non-string input as needed
    except:
        return 'unknown' # Return 'unknown' if language detection fails

# Apply the function to the 'judul' column and create a new column for detected language
df_cleaned['detected_language'] = df_cleaned['judul'].apply(detect_language)

# Filter out rows where the detected language is 'en' (English)
df_filtered = df_cleaned[df_cleaned['detected_language'] != 'en'].drop(columns=['detected_language'])


print("Original DataFrame shape:", df_cleaned.shape)
print("DataFrame shape after removing English titles:", df_filtered.shape)

display(df_filtered.head())
```

# Ngecek English titles

```python
# Filter for rows where the detected language is 'en'
df_removed_english = df_cleaned[df_cleaned['detected_language'] == 'en']

print("Rows removed (English titles):")
display(df_removed_english)
```

# Save as csv

```python
# Save the cleaned DataFrame to a new CSV file in Colab's local environment
output_file_path = 'data_link_berita_with_content_cleaned.csv' # You can change the path if needed

try:
    df_filtered.to_csv(output_file_path, index=False)
    print(f"Successfully saved cleaned data to {output_file_path}")
except Exception as e:
    print(f"An error occurred while saving the file: {e}")
```

# Data Description (Belum Praproses)

```python
# Perform descriptive statistical analysis
print("Descriptive Statistics:")
display(df_filtered.describe(include='all'))

# Display information about the DataFrame
print("\nDataFrame Info:")
display(df_filtered.info())

# Display the number of unique values in each categorical column
print("\nUnique values in categorical columns:")
for col in df_filtered.select_dtypes(include='object').columns:
    print(f"- {col}: {df_filtered[col].nunique()} unique values")

# Additional descriptive analysis based on data patterns

# Convert 'tanggal' column to datetime objects for easier analysis
df_filtered['tanggal'] = pd.to_datetime(df_filtered['tanggal'])

print("\nDistribution of 'portal':")
display(df_filtered['portal'].value_counts().head(10)) # Display top 10 portals

print("\nDistribution of 'tag':")
display(df_filtered['tag'].value_counts())

print("\nDate Range of Articles:")
print(f"Earliest Date: {df_filtered['tanggal'].min()}")
print(f"Latest Date: {df_filtered['tanggal'].max()}")
```
